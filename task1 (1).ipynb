```python
import pandas as pd
import numpy as np
df=pd.read_excel('student_performance(1).xlsx')
df.head()
```

```python
d=len(df)
print("total number of students :", d)
```

```python
df1=df[df['study_hours_per_day']>2]
print("number of students with study hours more than 2 : ", len(df1))
```

```python
df2=df[df['exam_score']>40.0]
print("number of students who passed : ", len(df2))
```

```python
studied_and_passed = df[(df['study_hours_per_day'] > 2) & (df['exam_score'] > 40.0)]
print("Students who studied > 2 hours and passed:", len(studied_and_passed))
```

```python
percentage=len(studied_and_passed)/len(df1)*100
print("percentage of students who studied more than 2 hours and passed : ", percentage)
```

```python
P1=len(df1)/d
P2=len(df2)/d
P3=len(studied_and_passed)/len(df1)

P_student_given_they_passed=P3*P1/P2
print("P(student>2|passed) : ", P_student_given_they_passed)
```
